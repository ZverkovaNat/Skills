Подключение к БД с помощью модуля sqlite3.

```python
import sqlite3
import pandas as pd
```

```python
con = sqlite3.connect('/content/book_simple.db.db')
cur = con.cursor()
```

# 1
Вывод авторов, имеющихся в базе.

```python
avtory = """
SELECT name AS Авторы
FROM author;
"""
pd.read_sql(avtory, con)
```

## 2

Поиск всех авторов с инициалами "С. А.". Автор в базе определяется так: И. О. Фамилия

```python
avtory_ca = """
SELECT name AS Авторы
FROM author
WHERE name LIKE 'С._А._%';
"""
pd.read_sql(avtory_ca, con)
```

## 3
Вывод списка книг, выпущенных в издательстве "Мир".


```python
reqest_mir = """
SELECT book.title,
       publisher.name
FROM book, publisher
WHERE publisher.name IS 'Мир';

"""
pd.read_sql(reqest_mir, con)
```

## 4

Вывод количества книг, принадлежащих каждой из категорий. Результат отсортирован по возрастанию.


```python
reqest_cat = """
SELECT COUNT(book.id) AS 'Количество книг', category.title AS 'Категория'
FROM book, category
WHERE (category.id = book.category_id)
GROUP BY category.title
ORDER BY COUNT(book.id) ASC;
"""
pd.read_sql(reqest_cat, con)
```

## 5
Вывод количества книг, выпущенных в каждом издательстве. В итоговом списке  только те издательства, которые выпустили 3 и более книг. Результат отсортирован по убыванию.


```python
reqest_izdat = """
SELECT COUNT(book.id) AS 'Количество книг', publisher.name AS 'Издательство'
FROM book, publisher
WHERE (publisher.id = book.publisher_id)
GROUP BY publisher.name
HAVING COUNT(book.id) >= 3
ORDER BY COUNT(book.id) DESC;
"""
pd.read_sql(reqest_izdat, con)
```

## 6

Вывод списка книг, автором которых являлся <<С. А. Айвазян>>.


```python
reqest_author = """
SELECT book.title, author.name
FROM book, authorbook, author
WHERE (authorbook.book_id = book.id) AND (authorbook.author_id = author.id) AND (author.name == 'С. А. Айвазян');
"""
pd.read_sql(reqest_author, con)
```

## 7

В каких издательствах и сколько выпускались книги за авторством <<С. А. Айвазян>>. Результат отсортирован по убыванию.

```python
reqest_last = """
SELECT author.name,
       COUNT(book.title),
       publisher.name
FROM book
INNER JOIN authorbook
ON book.id = authorbook.book_id
INNER JOIN author
ON author.id = authorbook.author_id
INNER JOIN publisher
ON publisher.id = book.publisher_id
WHERE author.name = 'С. А. Айвазян'
GROUP BY publisher.name
ORDER BY COUNT(book.id) DESC;
"""
pd.read_sql(reqest_last, con)
```
